```python
#Split  dataset into train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=123)
```

```python
import sklearn 
```

```python
from sklearn .linear_model import LinearRegression
```

```python
car=LinearRegression()
```

```python
car.fit(X_train,y_train)
```

```python
y_train_pred=car.predict(X_train)
y_test_pred=car.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
def model_performance(y_actual,y_pred):
    r2=r2_score(y_actual,y_pred)
    RMSE=np.sqrt(mean_squared_error(y_actual,y_pred))
    print("r2_score:{},RMSE :{}".format(round(r2,2),round(RMSE,2)))
```

```python
print("Train Performance")
model_performance(y_train,y_train_pred)
print("Train Performance")
model_performance(y_test,y_test_pred)
```

```python
from sklearn.linear_model import Ridge,Lasso
```

```python
l1=Lasso(alpha=9)
l1.fit(X_train,y_train)
y_pred_train=l1.predict(X_train)
y_pred_test=l1.predict(X_test)
```

```python
print("Train perfomance")
model_performance(y_train,y_pred_train)
print("Test perfomance")
model_performance(y_test,y_pred_test)
```

```python
for i in range(1,250):
    l2=Ridge(i)
    l2.fit(X_train,y_train)
    y_pred_train=l2.predict(X_train)
    y_pred_test=l2.predict(X_test)
    print(f"{i}:{l2.score(X_test,y_test)}")

```
