```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')
import plotly.express as px
```

```python
df=pd.read_csv("Air Pollution.csv",encoding='latin1')
```

```python
df.head(10)
```

```python
df.replace("-",np.nan,inplace=True)
```

```python
convert_dict = {'JAN': float,
                'FEB': float,
                'MAR': float,
                'APR': float,
                'MAY': float,
                'JUN': float,
                'JUL': float,
                'AUG': float,
                'SEP': float,
                'OCT': float,
                'NOV': float,
                'DEC': float,
                '2021': float,
                '2020': float,
                '2019': float,
                '2018': float,
                '2017': float 
                }
 
df = df.astype(convert_dict)
```

```python
df.drop(["2017","2018"],axis=1,inplace=True)
```

```python
df = df.fillna(df.median())
```

```python
def plots(x,y):
    plt.figure(figsize=(20,10))
    lineplot=sns.lineplot(x=df[x],y=df[y],data=df) 
    return(lineplot)
```

```python
def maxs(x):
    max_2021_polluted=df.groupby(x).max().tail(5)
    return max_2021_polluted
    
```

```python
def mins(x):
    min_2021_polluted=df.groupby(x).min().head(5)
    return min_2021_polluted
    
```

# which is most polluted city in year 2021 ?

```python
maxs("2021")
```
